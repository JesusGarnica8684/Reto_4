## Ejercicio 1:
Dado un número entero, determinar si ese número corresponde al código ASCII de una vocal minúscula.


```python
if __name__ == "__main__":
    n = int(input("Ingrese el numero que desea evaluar"))
    if n >= 97 and n <= 122: #se restringe el rango que puede adoptar n, segun el código ASCII
        print ("El numero corresponde a la minuscula ", chr(n), " en codigo ASCII.")
    else:
        print ("El numero no corresponde a una minúscula.")
```

## Ejercicio 2:
Dada una cadena de longitud 1, determine si el código ASCII de primera letra de la cadena es par o no.

```python
if __name__ == "__main__":
    charOne = input("Ingrese un caracter ")
    if len(charOne)==1: #se comprueba que el dato ingresado sea un string de 1 de longitud
        if ord(charOne) %2 == 0: #se verifica que si es divisible por dos es par
            print ("El caracter ", charOne, " es par, puesto su codigo ASCII corresponde a ", ord(charOne))
        else:
            print ("El caracter {charOne} es impar, puesto su codigo ASCII corresponde a ", ord(charOne))
    else:
        print ("Error, ingrese un caracter.")#Dado el caso se ingrese un string que no sea un caracter se regresa error
```

## Ejercicio 3:
Dado un carácter, construya un programa en Python para determinar si el carácter es un dígito o no.


```python
if __name__ == "__main__":
    charOne = input("Ingrese un caracter ")
    if len(charOne)==1: #se comprueba que el dato ingresado sea un string de 1 de longitud
        if charOne.isdigit: #se verifica que sea un digito
             print ("El caracter ", charOne, " es un digito")
        else:
            print ("El caracter ", charOne, " no es un digito")
    else:
        print ("Error, ingrese un caracter.")#Dado el caso se ingrese un string que no sea un caracter se regresa error
```

## Ejercicio 4
Realice un programa que lea dos números reales y determine si el primero es múltiplo del segundo

```python
if __name__ == "__main__":
    numPri = float(input("Ingrese el primer digito"))
    numSeg = float(input("Ingrese el segundo digito"))
    if numPri%numSeg == 0: #se verifica que sen multiplos por medio de su residuo, si es igual acero son multiplos
        print (numPri, " es multiplo de ", numSeg)
    else:
        print (numPri, " no es multiplo de ", numSeg)
```

## Ejercicio 5:
Dado un número real x, construya un programa que permita determinar si el número es positivo, negativo o cero. Para cada caso de debe imprimir el texto que se especifica a continuación:

* Positivo: "El número x es positivo"
* Negativo: "El número x es negativo"
* Cero (0): "El número x es el neutro para la suma"

```python
if __name__ == "__main__":
    numX = float(input("Ingrese el número a evaluar:"))
    if numX > 0: #se evalua que sea mayor a cero, o sea positivo
        print("El número ", numX," es positivo")
    elif numX < 0: #se evalua que sea menor a cero, o sea negativo.
        print("El número ", numX," es negativo")
    else: #en su defecto el numero es cero.
        print("El número ", numX," es el neutro para la suma")
```

## Ejercicio 6:
Dado el centro y el radio de un círculo, determinar si un punto de R2 pertenece o no al interior del círculo.

```python
if __name__ == "__main__":
    Xcentro = int(input("Digite el valor de X en el centro del círculo"))
    Ycentro = int(input("Digite el punto Y del centro de la circunferencia"))
    rad = int(input("Digite el radio de la circunferencia"))
    Xn = int(input("Digite el componente en X del punto que desea verificar"))
    Yn = int(input("Digite el componente de y del punto a verificar"))

    if (Xcentro - Xn)**2 + (Ycentro - Yn)**2 <= rad**2: #se utiliza la ecuación canónica de la circunferencia,
        #por medio de esta si los valores son menores o iguales al radio, está dentro de la circunferencia
        print("el punto (", Xn, ",", Yn, ") pertenece a la circunferencia")
    else:
        print("el punto (", Xn, ",", Yn, ") no pertenece a la circunferencia")
```

## Ejercicio 7:
Dadas tres longitudes positivas, determinar si con esas longitudes se puede construir un triángulo.

```python
if __name__ == "__main__":
    print ("Para construir un triangulo se necesita la longitud de los tres lados.")
    ladoUno = int(input("Ingrese una longitud positiva para el primer lado: "))
    ladoDos = int(input("Ingrese una longitud positiva para el segundo lado: "))
    ladoTres = int(input("Ingrese una longitud positiva para el tercer lado: "))
    if ladoUno > 0 and ladoDos > 0 and ladoTres > 0: #se verifica que las longitudes sean positivas y mayores a cero
        if ladoUno + ladoDos > ladoTres: #Se puede formar un triangulo si la suma de dos de sus lados es mayor a su tercer lado
            print("Se puede construir un triangulo.")
        else:
            print("No se puede construir un triangulo. ")
    else:
        print ("Error en ingreso de datos. Las longitudes deben ser positivas")
```
